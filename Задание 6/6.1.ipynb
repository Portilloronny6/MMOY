```python
# read the students_Data.csv
import pandas as pd

students_df = pd.read_csv('students_Data.csv')

students_df.head()
```

```python
# Apply the KFold strategy with division into 5 blocks without shuffling. Return lists of indices of objects included in the test and training sets at each iteration, and form the actual test and training sets (classes and data labels).
from sklearn.model_selection import KFold


kf = KFold(n_splits=5, shuffle=False)


for train_index, test_index in kf.split(students_df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = students_df.iloc[train_index], students_df.iloc[test_index]
    y_train, y_test = students_df.iloc[train_index], students_df.iloc[test_index]
    print("X_train:\n", X_train)
    print("X_test:\n", X_test)
    print("y_train:\n", y_train)
    print("y_test:\n", y_test)
    print("\n")
```

```python
# Apply the KFold strategy by dividing into 5 blocks with shuffling. Return lists of indices of objects included in the test and training sets at each iteration, and form the actual test and training sets (classes and data labels).
from sklearn.model_selection import KFold


kf = KFold(n_splits=5, shuffle=True)


for train_index, test_index in kf.split(students_df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = students_df.iloc[train_index], students_df.iloc[test_index]
    y_train, y_test = students_df.iloc[train_index], students_df.iloc[test_index]
    print("X_train:\n", X_train)
    print("X_test:\n", X_test)
    print("y_train:\n", y_train)
    print("y_test:\n", y_test)
    print("\n")
```

```python
# Apply the StratifiedKFold strategy with 5 blocks. Show lists of indices of objects included in the test and training sets at each iteration, and form the actual test and training sets (classes and data labels).
from sklearn.model_selection import StratifiedKFold


kf = StratifiedKFold(n_splits=5)


for train_index, test_index in kf.split(students_df, students_df['Задолженность']):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = students_df.iloc[train_index], students_df.iloc[test_index]
    y_train, y_test = students_df.iloc[train_index], students_df.iloc[test_index]
    print("X_train:\n", X_train)
    print("X_test:\n", X_test)
    print("y_train:\n", y_train)
    print("y_test:\n", y_test)
    print("\n")
```

```python
# Apply the ShuffleSplit strategy with 10 block splits. Return lists of indices of objects included in the test and training sets at each iteration, and form the actual test and training sets (classes and data labels).
from sklearn.model_selection import ShuffleSplit


kf = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)


for train_index, test_index in kf.split(students_df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = students_df.iloc[train_index], students_df.iloc[test_index]
    y_train, y_test = students_df.iloc[train_index], students_df.iloc[test_index]
    print("X_train:\n", X_train)
    print("X_test:\n", X_test)
    print("y_train:\n", y_train)
    print("y_test:\n", y_test)
    print("\n")
```

```python
# Apply the StratifiedShuffleSplit strategy with 10 split blocks. Return lists of indices of objects included in the test and training sets at each iteration, and form the actual test and training sets (classes and data labels).
from sklearn.model_selection import StratifiedShuffleSplit


kf = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=0)


for train_index, test_index in kf.split(students_df, students_df['Задолженность']):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = students_df.iloc[train_index], students_df.iloc[test_index]
    y_train, y_test = students_df.iloc[train_index], students_df.iloc[test_index]
    print("X_train:\n", X_train)
    print("X_test:\n", X_test)
    print("y_train:\n", y_train)
    print("y_test:\n", y_test)
    print("\n")
```
