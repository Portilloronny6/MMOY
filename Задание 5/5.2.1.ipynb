```python
import random
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification

random_state = random.randint(0, 500)
X, y = make_classification(n_samples=200, n_features=2, n_informative=1, n_redundant=0, n_classes=2,
                           n_clusters_per_class=1, random_state=random_state, flip_y=5.5)
```

```python
from matplotlib.colors import ListedColormap

colors = ListedColormap(['red', 'blue'])
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=colors)
```

```python
# Experiment with the flip_y value: set the corresponding parameter values to 4.0, 7.0 and 10.0, display graphs in a row with headers reporting the noise parameter value used.


def make_noise_classification(noise):
    return make_classification(n_samples=200, n_features=2, n_informative=1, n_redundant=0, n_classes=2,
                               n_clusters_per_class=1, random_state=random_state, flip_y=noise)


for noise in [4.0, 7.0, 10.0]:
    X, y = make_noise_classification(noise)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=colors)
    plt.title('Noise: {}'.format(noise))
    plt.show()
```
