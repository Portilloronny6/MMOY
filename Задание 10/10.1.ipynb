```python
# Using the datasets module of the Scikit-learn library, generate a model dataset for a multiclass classification problem with two informative features and three classes. Ensure reproducible results.
import random
from sklearn.datasets import make_multilabel_classification

random_state = random.randint(0, 500)
X, y = make_multilabel_classification(
    n_samples=1000, n_classes=3, n_labels=2,
    allow_unlabeled=False, random_state=random_state
)
```

```python
# Visualize the generated point clouds.
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=y[:, 0], s=40, cmap='viridis')
plt.show()
```

```python
# Split the data set into training and test samples in the ratio 70/30.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_state)
```

```python
# Create a decision tree model with default parameters (ensure only reproducibility of results) and train it on the training set.
from sklearn.tree import DecisionTreeClassifier, plot_tree

tree_clf = DecisionTreeClassifier(random_state=random_state)
tree_clf.fit(X_train, y_train)
```

```python
# Get the predictions of the trained model for the objects of the test set. Evaluate the quality of classification using the accuracy metric both on the training and test sets.
y_pred = tree_clf.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
# Create a decision tree model with a depth of 1 and train it on the training set. Get model-predicted class labels for training and test set objects. Evaluate the quality of classification using the accuracy metric on the training and test sets.
tree_clf = DecisionTreeClassifier(max_depth=1, random_state=random_state)
tree_clf.fit(X_train, y_train)

y_pred = tree_clf.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
# graph the decision tree model.
plot_tree(tree_clf, filled=True)
```

```python
# Create a decision tree model with a depth of 2 and execute with it
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=random_state)
tree_clf.fit(X_train, y_train)
```

```python
# graph the decision tree model.
plot_tree(tree_clf, filled=True)
```

```python
# Create a decision tree model with a depth of 3 and execute with it
tree_clf = DecisionTreeClassifier(max_depth=3, random_state=random_state)
tree_clf.fit(X_train, y_train)
```

```python
# graph the decision tree model.
plot_tree(tree_clf, filled=True)
```

```python
# Create a decision tree model with a limit on the number of objects in a sheet (specify no more than 3 objects). Train this model on the training set. Get model-predicted class labels for training and test set objects. Evaluate the classification quality using the accuracy metric on the training and test sets.
tree_clf = DecisionTreeClassifier(max_leaf_nodes=3, random_state=random_state)
tree_clf.fit(X_train, y_train)

y_pred = tree_clf.predict(X_test)

accuracy_score(y_test, y_pred)
```
